```python
#import library BeautifulSoup
from bs4 import BeautifulSoup
import requests

#mengambil URL
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#mencari tabel berdasarkan class
table = soup.find('table', class_ = 'wikitable sortable')
print(table)
```

```python
#mencari judul tabel dengan tag (th)
world_tittles = table.find_all('th')
world_tittles
```

```python
#merapikan judul tabel
table_world_tittles = [tittle.text.strip() for tittle in world_tittles]
print(table_world_tittles)
```

```python
#import library pandas dan membuat data frame untuk judul
import pandas as pd
df = pd.DataFrame(columns = table_world_tittles)
df
```

```python
#mencari kolom pada tabel
column_data = table.find_all('tr')

#looping untuk memanggil baris pada setiap kolom dengan tag (td)
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    #memasukkan dan menyesuaikan data tabel dengan df(data frame)
    length = len(df)
    df.loc[length] = individual_row_data
    
df
```

```python
#memasukkan data pada df (data frame) ke dalam file excel (csv)
df.to_csv(r'E:\DATA KULIAH\MY FILE\Python Project\Companies.csv', index = False)
```
